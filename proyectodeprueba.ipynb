```python
df_blancos = df[
    (df['species'].str.lower() == 'white shark') |
    (df['species'].str.lower() == 'great white shark')
] 

df_blancos['species'] = 'tiburón blanco'

```

```python
especies_unicas = df['species'].dropna().unique()
```

```python
palabras_clave = ['white', 'blanco', 'branco', 'requin', 'great white', 'carcharodon carcharias', 'great white shark', 'white shark']
```

```python
df_tiburon_blanco = df[df['species'].str.contains('|'.join(palabras_clave), na=False)]
```

```python
df_tiburon_blanco
```

```python
df_tiburon_blanco.drop('location', axis =1, inplace=True)
df_tiburon_blanco.drop('name', axis =1, inplace=True)
df_tiburon_blanco.drop('pdf', axis =1, inplace=True)
df_tiburon_blanco.drop('href formula', axis =1, inplace=True)
df_tiburon_blanco.drop('href', axis =1, inplace=True)
df_tiburon_blanco.drop('case number', axis =1, inplace=True)
df_tiburon_blanco.drop('case number.1', axis =1, inplace=True)
df_tiburon_blanco.drop('original order', axis =1, inplace=True)
df_tiburon_blanco.drop('unnamed: 21', axis =1, inplace=True)
df_tiburon_blanco.drop('unnamed: 22', axis =1, inplace=True)
```
